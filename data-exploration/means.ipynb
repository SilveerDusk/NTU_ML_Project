```python
import pandas as pd
import numpy as np

# Step 1: Read the data
data = pd.read_csv('../data-for-stage-1/train_data_copy_jason.csv')

# Step 2: Define the home and away data
data.columns.tolist()
home_columns = []
away_columns = []
others_columns = []
for col in data.columns:
  if 'abbr' in col or 'win' in col:
    pass
  elif 'home' in col:
    home_columns.append(col)
  elif 'away' in col:
    away_columns.append(col)
  else:
    others_columns.append(col)

print(home_columns)
print(away_columns)
    
# Step 3: Filter the data
data = data.filter(items = [
    "home_team_abbr",
    "away_team_abbr",
    "home_team_win",
    ] + home_columns + away_columns)

zero_rows = 0

# Step 5 find the means of the data
for team in data['home_team_abbr'].drop_duplicates():
  print('Updating', team, '...')
  team_data = data[data['home_team_abbr'] == team].copy()

  for feat in home_columns:
    print('Updating', feat, '...')
    sum = 0
    non_zeros = 1

    for index, value in team_data[feat].items():
      if (type(value) is not float) or (pd.isna(value)):
         team_data.loc[index, feat] = sum / non_zeros
      else:
        sum += value
        non_zeros += 1

    if sum == 0:
      print('All zeros for', feat, 'in', team)
      zero_rows += 1
      all_teams_mean = data[feat].mean()
      if all_teams_mean == 0:
        print('All zeros for', feat, 'in all teams. What The FUCK!')
      else:
        team_data[feat].ffill()

      
  for feat in away_columns:
    print('Updating', feat, '...')
    sum = 0
    non_zeros = 1

    for index, away_game in team_data[feat].items():
      if (type(value) is not float) or (pd.isna(value)):
         team_data.loc[index, feat] = sum / non_zeros
      else:
        sum += value
        non_zeros += 1

    if sum == 0:
      print('All zeros for', feat, 'in', team)
      zero_rows += 1

  data[data['home_team_abbr'] == team] = team_data


data.to_csv("running_means_training.csv", index=False)

print('Zero rows:', zero_rows)
```
